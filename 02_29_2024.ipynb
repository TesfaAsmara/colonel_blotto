```python
def generate_combinations(b_list, k_list, current_combination=[], all_combinations=[]):
    # Base case: when the current combination has reached the size of b_list or k_list
    if len(current_combination) == len(b_list):
        all_combinations.append(tuple(current_combination))
        return
    
    # The index to work on
    index = len(current_combination)
    
    # For the current pair (bi, ki), generate all comparisons
    for comparison in ['>', '=', '<']:
        # Append the current comparison and recurse
        next_combination = current_combination + [f"b{index+1} {comparison} k{index+1}"]
        generate_combinations(b_list, k_list, next_combination, all_combinations)


n = 3  # Size of the b and k lists
b_list = [f"b{i}" for i in range(1, n+1)]
k_list = [f"k{i}" for i in range(1, n+1)]

all_combinations = []
generate_combinations(b_list, k_list, [], all_combinations)

# Print or process the generated combinations
for combo in all_combinations:
    print(combo)
```

```python
def process_combinations(all_combinations):
    processed_combinations = []

    for combo in all_combinations:
        processed_combo = []
        for i, entry in enumerate(combo):
            if '>' in entry:
                processed_combo.append(i+1)  # Positive its index for '>'
            elif '=' in entry:
                processed_combo.append(0)  # Zero for '='
            elif '<' in entry:
                processed_combo.append(-i-1)  # Negative its index for '<'
        processed_combinations.append(tuple(processed_combo))
    
    return processed_combinations


processed_combinations = process_combinations(all_combinations)

# Print or process the processed combinations
for combo in processed_combinations:
    print(combo)
```

```python
def sum_processed_combinations(processed_combinations):
    # Majoritarian objective, no sweeping mechanism
    sums = [sum(combo) for combo in processed_combinations]
    return sums


sums = sum_processed_combinations(processed_combinations)

# Mapping sums back to original comparison strings
def map_sums_to_comparisons(sums, all_combinations):
    # Create a mapping of sum to combinations
    sum_to_combinations = {}
    for combo_sum, combo in zip(sums, all_combinations):
        if combo_sum not in sum_to_combinations:
            sum_to_combinations[combo_sum] = []
        sum_to_combinations[combo_sum].append(combo)
    
    return sum_to_combinations

sum_to_combinations = map_sums_to_comparisons(sums, all_combinations)

# Print the mapping
for combo_sum, combos in sum_to_combinations.items():
    print(f"Sum: {combo_sum}")
    for combo in combos:
        print(combo)
```

```python
def count_comparison_operators(sum_to_combinations):
    # Initialize a dictionary to hold counts of each operator in each position
    position_counts = {}
    
    # Filter and work only with positive sums
    for combo_sum, combos in sum_to_combinations.items():
        if combo_sum > 0:
            for combo in combos:
                for i, comparison in enumerate(combo):
                    # Initialize position i if not already
                    if i not in position_counts:
                        position_counts[i] = {'>': 0, '=': 0, '<': 0}
                    
                    # Identify the comparison operator and increment its count
                    if '>' in comparison:
                        position_counts[i]['>'] += 1
                    elif '=' in comparison:
                        position_counts[i]['='] += 1
                    elif '<' in comparison:
                        position_counts[i]['<'] += 1
    
    return position_counts

# Count the comparison operators for combinations with positive sums
position_counts = count_comparison_operators(sum_to_combinations)

# Print the counts
for position, counts in position_counts.items():
    print(f"Position {position + 1}:")
    for comparison, count in counts.items():
        print(f"  {comparison}: {count}")
```

```python
def count_positions(combinations):
    position_counts = [{'>': 0, '=': 0, '<': 0} for _ in range(len(combinations[0]))]
    
    for combo in combinations:
        for i, comparison in enumerate(combo):
            if '>' in comparison:
                position_counts[i]['>'] += 1
            elif '=' in comparison:
                position_counts[i]['='] += 1
            elif '<' in comparison:
                position_counts[i]['<'] += 1
    
    return position_counts

def calculate_percentages(positive_counts, total_counts):
    percentages = []
    for pos in range(len(total_counts)):
        percentages.append({
            '>': (positive_counts[pos]['>'] / total_counts[pos]['>'] * 100) if total_counts[pos]['>'] else 0,
            '=': (positive_counts[pos]['='] / total_counts[pos]['='] * 100) if total_counts[pos]['='] else 0,
            '<': (positive_counts[pos]['<'] / total_counts[pos]['<'] * 100) if total_counts[pos]['<'] else 0,
        })
    return percentages


total_counts = count_positions(all_combinations)

# Filter combinations with a positive sum
positive_combinations = [combo for combo, sum_ in zip(all_combinations, sums) if sum_ > 0]

# Count occurrences in positive combinations
positive_counts = count_positions(positive_combinations)

# Calculate percentages
percentages = calculate_percentages(positive_counts, total_counts)

# Print the percentages
for i, pos_percentages in enumerate(percentages, 1):
    print(f"Position {i}:")
    for comparison, percentage in pos_percentages.items():
        print(f"  {comparison}: {percentage:.2f}%")
```

```python
import os
import json
import itertools


def sums(length, total_sum):
    if length == 1:
        yield (total_sum,)
    else:
        for value in range(total_sum + 1):
            for permutation in sums(length - 1, total_sum - value):
                yield (value,) + permutation
```
